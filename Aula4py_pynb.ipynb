```python

import numpy as np

df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Diana', 'Eduardo'],
    'salario': [4000, np.nan, 3500, np.nan, 5000]
})

# Preencher com a mdia salarial
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

# Preencher com a mediana salarial
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['Seg', 'Ter', 'Qua', 'Qui', 'Sex'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['Seg', 'Ter', 'Qua', 'Qui', 'Sex'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Diana', 'Eduardo'],
    'cidade': ['So Paulo', np.nan, 'Curitiba', np.nan, 'Salvador']
})

df_cidades['cidade_corrigida'] = df_cidades['cidade'].fillna('No informado')

df_cidades
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x='experience_level', y='salary_in_usd')
plt.title("Salrio mdio por senioridade")
plt.xlabel("Senioridade")
plt.ylabel("Salrio mdio anual (USD)")
plt.show()
```

```python

plt.figure(figsize=(10,5))
sns.histplot(df_limpo['salary_in_usd'], bins = 50, kde=True)
plt.title("Distribuio dos salrios anuais")
plt.xlabel("Salrio em USD")
plt.ylabel("Frequncia")
plt.show()
```

```python

plt.figure(figsize=(8,5))
sns.boxplot(x=df_limpo['salary_in_usd'])
plt.title("Boxplot Salrio")
plt.xlabel("Salrio em USD")
plt.show()
```

```python
ordem_senioridade = ['EN', 'MI', 'SE', 'EX']

plt.figure(figsize=(8,5))
sns.boxplot(x='experience_level', y='salary_in_usd', data=df_limpo, order=ordem_senioridade, palette='Set2', hue='experience_level')
plt.title("Boxplot da distribuio por senioridade")
plt.xlabel("Salrio em USD")
plt.show()
```

```python

import plotly.express as px
```

```python
senioridade_media_salario = df_limpo.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(senioridade_media_salario,
             x='experience_level',
             y='salary_in_usd',
             title='Mdia Salarial por Senioridade',
             labels={'experience_level': 'Nvel de Senioridade', 'salary_in_usd': 'Mdia Salarial Anual (USD)'})

fig.show()
```

```python
remoto_contagem = df_limpo['remote_ratio'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Proporo dos tipos de trabalho'

          )

fig.show()
```

```python
remoto_contagem = df_limpo['remote_ratio'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Proporo dos tipos de trabalho',
             hole=0.5
          )
fig.update_traces(textinfo='percent+label')

fig.show()
```
